```python
class Book:
    def __init__(self, title, author, ISBN):
        self.title = title
        self.author = author
        self.ISBN = ISBN
        self.is_checked_out = False

    def check_out(self):
        if not self.is_checked_out:
            self.is_checked_out = True
            return f"{self.title} by {self.author} has been checked out."
        else:
            return f"{self.title} by {self.author} is already checked out."

    def check_in(self):
        if self.is_checked_out:
            self.is_checked_out = False
            return f"{self.title} by {self.author} has been checked in."
        else:
            return f"{self.title} by {self.author} is not currently checked out."

    def __str__(self):
        return f"{self.title} by {self.author}"


class Patron:
    def __init__(self, name):
        self.name = name
        self.checked_out_books = []

    def check_out_book(self, book):
        message = book.check_out()
        if "checked out" in message:
            self.checked_out_books.append(book)
        return message

    def return_book(self, book):
        message = book.check_in()
        if "checked in" in message:
            self.checked_out_books.remove(book)
        return message

    def list_checked_out_books(self):
        if self.checked_out_books:
            books_str = ", ".join(str(book) for book in self.checked_out_books)
            return f"Books checked out by {self.name}: {books_str}"
        else:
            return f"{self.name} has no books checked out."


# Example usage:
if __name__ == "__main__":
    # Creating books
    book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "9780142437344")
    book2 = Book("To Kill a Mockingbird", "Harper Lee", "9780061120084")

    # Creating patrons
    patron1 = Patron("Alice")
    patron2 = Patron("Bob")

    # Checking out and returning books
    print(patron1.check_out_book(book1))
    print(patron2.check_out_book(book1))
    print(patron1.list_checked_out_books())

    print(patron1.return_book(book1))
    print(patron2.return_book(book2))
    print(patron1.list_checked_out_books())
```
